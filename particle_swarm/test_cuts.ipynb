```python
def print_cuts(gm, A, B):
    print('-- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- ')
    print('cut      = ' + str(gm.cut_cost(A, B)))
    print('norm_cut = ' + str(gm.cut_normalised_cost(A, B)))
    print('cond     = ' + str(gm.cut_conductance(A, B)))
    print('e_exp    = ' + str(gm.cut_edge_expansion(A, B)))
    print('m_exp    = ' + str(gm.cut_mixing_expansion(A, B)))
    print('-- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- ')
```
